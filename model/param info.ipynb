```python
import pandas as pd
import numpy as np
import pickle 
import json
```

```python
b = ['annealing', 'BulkModulus','Density','Littleton','ND' ,'Shear_modulus','softening','TEC','TG','TLiq','YM','Hardness_RT']
path = r'/chalco_results2' 
df = pd.DataFrame()
# init = 'annealing'
for prop in b:
    output = path+ '/' + prop + '/'
    file = output  + prop+'_XGB2'+'.pickle'
    ls = pd.read_pickle(file)
    param = ls[3]
    i = prop
    if i == 'annealing':
        lab = "Annealing point (K) " 

    if i == 'BulkModulus':
        lab = "Bulk modulus (GPa) "
    if i == 'Density':
        lab = "Density (g/cm$^3$) "
    if i == 'Hardness_RT' or i == 'Hardness_misc':
        lab ="Vickers hardness (GPa) "
  
    if i== 'shear_modulus':
        lab = "Shear modulus (GPa) "
    if i== 'softening':
        lab = "Softening point (K) "
    if i== 'TLiq':
        lab = "Liquidus temperature (K) "
    if i == 'Littleton':
        lab = "Littleton point (K) "
    if i == 'TG':
        lab = "$T_g$ (K) "
    if i== 'TEC':
        lab ="TEC, \u03B1  (ln[$10^{-7}$ $K^{-1}$]) "
    if i== 'YM':
        lab = "Young's modulus (GPa) "
    if i== 'ND':
        lab =  "Refractive index, $n_d$ "


    param["Property"] = prop
    param = pd.DataFrame([param])
    df = pd.concat([df, pd.DataFrame(param, columns= param.columns )], ignore_index=True)
#     for prop in b:
#     output = path+ '/' + prop + '/'
#     file = output  + prop+'_XGB2'+'.pickle'
#     ls = pd.read_pickle(file)
#     param = ls[3]
```
